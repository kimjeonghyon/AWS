```python
import autogluon
import pandas as pd
```

```python
#Data Source: https://www.kaggle.com/datasets/arunjangir245/boston-housing-dataset
data = pd.read_csv('BostonHousing.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
#Checking for missing values
data.isnull().sum()
```

```python
data['rm']=data['rm'].fillna(data['rm'].mean())
data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(data, test_size=0.1)
train.to_csv('housing_train.csv', index=False)
test.to_csv('housing_test.csv', index=False)
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor

train_data = TabularDataset('housing_train.csv')
train_data.head()
```

```python
predictor = TabularPredictor(label='medv',path='model').fit(train_data)
```

```python
# !pip install bokeh==2.0.1
```

```python
results = predictor.fit_summary()

results
```

```python
# 예측 전처리

test_data = TabularDataset('housing_test.csv')
y_test = test_data['medv']
test_data_nolab = test_data.drop(columns=['medv'])
test_data_nolab.head()
```

```python
# 예측 실행
# predictor = TabularPredictor.load('model')
y_pred = predictor.predict(test_data_nolab)
print('predictions: \n", ', y_pred)
```

```python
#모델 평가
perf = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)
```

```python
#모델별 정확도
predictor.leaderboard(test_data, silent=True)
```

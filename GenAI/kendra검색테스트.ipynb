```python
from langchain.retrievers import AmazonKendraRetriever
```

```python
# KENDRA_INDEX_ID = "cf463aa7-0634-45e0-9c4e-52c3a0bb29bb"
KENDRA_INDEX_ID = "cf463aa7-0634-45e0-9c4e-52c3a0bb29bb"
```

```python
bedrock_region ='us-east-1'
model_id = 'anthropic.claude-v2'
print('bedrock_region: ', bedrock_region)
print('model_id: ', model_id)
```

```python
# 속성 filter 조건
kendra_attribute_filter = {"AndAllFilters": [{
                        "EqualsTo": {      
                            "Key": "_language_code",
                            "Value": {
                                "StringValue": "ko"
                                }
                            }
                        }
                    ]}
#문서 추출 개수.
kendra_top_k = 5
```

```python
text = "Google App Engine은 뭐니?"
```

```python
kendra_retriever = AmazonKendraRetriever(index_id=KENDRA_INDEX_ID,
                                         attribute_filter=kendra_attribute_filter,
                                         top_k=kendra_top_k)
kendra_response = kendra_retriever.get_relevant_documents(text)
print('kendra_response: ', kendra_response)
```
